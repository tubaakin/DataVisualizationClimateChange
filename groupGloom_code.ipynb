```python
import pandas as pd
import matplotlib.pyplot as plt

k1 = pd.read_csv('sealevel.csv')
k1["year"] = pd.to_datetime(k1["Time"]).dt.year

meanGMSL = []
years = []
meanUncertainty = []

for i in range(1880,2013):
    meanGMSL.append ((k1[k1["year"] == i]["GMSL"].sum()) / (len(k1[k1["year"] == i])))
    meanUncertainty.append((k1[k1["year"] == i]["GMSL uncertainty"].sum()) / (len(k1[k1["year"] == i])))
    years.append(i)

gmslSet = pd.DataFrame({"gmsl":meanGMSL,"uncertainty":meanUncertainty,"year":years})

plt.plot(gmslSet["year"], gmslSet["uncertainty"], color='r', label='sin')
plt.plot(gmslSet["year"], gmslSet["gmsl"], color='g', label='cos')
plt.xlabel("Years")
plt.ylabel("Magnitude")
plt.legend(["GMSL Uncertainty", "GMSL"])
plt.title("GMSL and GMSL Uncertainty by Year")

plt.show()
```
